```python
import nltk
import gensim.models.word2vec as w2v
from sklearn.manifold import TSNE
import re
import matplotlib.pyplot as plt
```

# Word2Vec on A Song of Ice and Fire

```python
got_book_1 = 'got1.txt'
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
type(stop_words)
```

```python
def clean(word):
    word = word.strip()
    word = word.lower()
    word = re.sub('[^A-Za-z0-9]+', '', word)
    if word not in stop_words:
        return word
    else:
        return ''
```

```python
line_count = 0
sentences = []

with open(got_book_1,'r',encoding="utf8") as inpFile:
    x = inpFile.readlines()
    for line in x:
        if line is not None or line != '\n':
            words = line.split()
            words = map(lambda x: clean(x), words)
            words = list(filter(lambda x:True if len(x) > 0 else False, words))
            if len(words)>0:
                sentences.append(words)
```

```python
sentences[1000:1100]
```

```python
w2v_model = w2v.Word2Vec(sentences, window=5, size=500, workers=4, min_count=5)
```

```python
words = list(w2v_model.wv.vocab)
words
```

```python
w2v_model.save('got_model.bin')
```

```python
loaded_model = w2v.Word2Vec.load('got_model.bin')
print(loaded_model)
```

# Visualize Word Embeddings

```python
X = loaded_model[loaded_model.wv.vocab]
```

```python
tsne = TSNE(n_components=2, random_state=0)
result = tsne.fit_transform(X)
```

```python
plt.scatter(result[:, 0], result[:, 1])
words = list(loaded_model.wv.vocab)
for i, word in enumerate(words):
	plt.annotate(word, xy=(result[i, 0], result[i, 1]))
```
